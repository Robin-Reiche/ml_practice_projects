# Daten aufteilen

```python
X = df.copy()
y = X.pop("Class")
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

```python
y.value_counts() / len(y)
```

```python
y_train.value_counts() / len(y_train)
```

```python
y_test.value_counts() / len(y_test)
```

# Hyperparameteroptimierung


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

model = DecisionTreeClassifier(random_state=0)
param_grid = {
    'max_depth': [7, 8, 9, 10, 11, 12]
}

grid_search = GridSearchCV(model, param_grid, cv=4, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best Accuracy: ", grid_search.best_score_)
print("Best Parameter: ", grid_search.best_params_)
```

# Accuracy

```python
final_model = grid_search.best_estimator_
y_predict = final_model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_predict)
print("Final Accuracy:", accuracy)
```

# Confusion Matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay

cm = ConfusionMatrixDisplay.from_predictions(y_test, y_predict)
cm.ax_.tick_params(axis='both', labelsize=7)
cm
```

# Precision, Recall und F1-Score (Klassenspezifisch)

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precisions = precision_score(y_test, y_predict, average=None)
print("Precisions: ", precisions)
```

```python
recalls = recall_score(y_test, y_predict, average=None)
print("Recalls: ", recalls)
```

```python
f1 = f1_score(y_test, y_predict, average=None)
print("F1-Scores: ", f1)
```

# Precision, Recall und F1-Score (Macro-Average)

```python
macro_precision = precision_score(y_test, y_predict, average="macro")
print("Macro-Precision: ", macro_precision)
```
