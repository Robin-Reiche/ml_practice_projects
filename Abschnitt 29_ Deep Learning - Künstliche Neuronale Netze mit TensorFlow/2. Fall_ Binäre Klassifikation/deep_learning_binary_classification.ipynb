# Daten aufteilen

```python
X = df.copy()
y = X.pop("Potability")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=0)
```

# Neuronales Netz erstellen (BinÃ¤re Klassifikation)

```python
import tensorflow as tf
from tensorflow.keras import layers

model_1 = tf.keras.Sequential([
    layers.Dense(units=9, activation="relu", input_shape=[9]),
    layers.Dense(units=9, activation="relu"),
    layers.Dense(units=1, activation="sigmoid"),
])

model_1.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()

pipeline_1 = Pipeline(steps=[
    ("scaler", standard_scaler),
    ("model", model_1)
])
```

```python
pipeline_1.fit(X_train, y_train, model__batch_size=512, model__epochs=50)
```

```python
pipeline_1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

y_predict_1 = pipeline_1.predict(X_test) >= 0.5
accuracy = accuracy_score(y_predict_1, y_test)

print("Accuracy:", accuracy)
```

# Hyperparameteroptimierung

```python
def create_neural_network(hidden_layers, hidden_layer_neurons, input_shape, hidden_layer_activation,
                 output_neurons, output_activation, optimizer, loss):
    model = tf.keras.Sequential()
    model.add(layers.Dense(units=hidden_layer_neurons, activation=hidden_layer_activation, input_shape=input_shape))

    for _ in range(hidden_layers - 1):
        model.add(layers.Dense(units=hidden_layer_neurons, activation=hidden_layer_activation))

    model.add(layers.Dense(units=output_neurons, activation=output_activation))
    model.compile(optimizer=optimizer, loss=loss)
    return model
```

```python
model_1 = create_neural_network(hidden_layers=2, hidden_layer_neurons=9, input_shape=[9], hidden_layer_activation="relu",
                 output_neurons=1, output_activation="sigmoid", optimizer="adam", loss="binary_crossentropy")
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

model_2 = KerasClassifier(model=create_neural_network)


pipeline_2 = Pipeline(steps=[
    ("scaler", standard_scaler),
    ("classifier", model_2)
])

param_grid = {
    "classifier__model__hidden_layers": [2, 3],
    "classifier__model__hidden_layer_neurons": [9, 18],
    "classifier__model__input_shape": [[9]],
    "classifier__model__hidden_layer_activation": ["relu"],
    "classifier__model__output_neurons": [1],
    "classifier__model__output_activation": ["sigmoid"],
    "classifier__model__optimizer": ["adam"],
    "classifier__model__loss": ["binary_crossentropy"],
}

grid_search = GridSearchCV(pipeline_2, param_grid, cv=4, scoring="accuracy")
grid_search.fit(X_train, y_train, classifier__batch_size=512, classifier__epochs=50)
```

```python
print("Best Accuracy: ", grid_search.best_score_)
print("Best Parameters: ", grid_search.best_params_)
```

# Early Stopping, um *epochs* besser zu bestimmen

```python
early_stopping_model = create_neural_network(hidden_layers=3, hidden_layer_neurons=9, input_shape=[9], hidden_layer_activation="relu",
                 output_neurons=1, output_activation="sigmoid", optimizer="adam", loss="binary_crossentropy")

early_stopping_pipeline = Pipeline(steps=[
    ("scaler", standard_scaler),
    ("model", early_stopping_model)
])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

X_train_new, X_valid, y_train_new, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
early_stopping = EarlyStopping(min_delta=0.0001, patience=50)

standard_scaler.fit(X_train_new)

early_stopping_pipeline.fit(
    X_train_new, y_train_new,
    model__validation_data=(standard_scaler.transform(X_valid), y_valid),
    model__batch_size=512,
    model__epochs=50000,
    model__callbacks=[early_stopping]
)
```

# Besten Wert fÃ¼r *epochs* ausgeben:

```python
best_number_of_epochs = early_stopping.best_epoch + 1
best_number_of_epochs
```

# Ergebnisse visualisieren

```python
import seaborn as sns

history = early_stopping_pipeline.named_steps["model"].history.history
history_df = pd.DataFrame(history)
history_df
```

```python
sns.lineplot(history_df)
```

# Finales neuronales Netz

```python
final_model = create_neural_network(hidden_layers=3, hidden_layer_neurons=18, input_shape=[9], hidden_layer_activation="relu",
                 output_neurons=1, output_activation="sigmoid", optimizer="adam", loss="binary_crossentropy")

final_pipeline = Pipeline(steps=[
    ("scaler", standard_scaler),
    ("model", final_model)
])

final_pipeline.fit(X_train, y_train, model__batch_size=512, model__epochs=best_number_of_epochs)
```

# Finales Ergebnis (Accuracy)

```python
y_predict_final = final_pipeline.predict(X_test) >= 0.5
accuracy = accuracy_score(y_test, y_predict_final)

print("Final Accuracy:", accuracy)
```
