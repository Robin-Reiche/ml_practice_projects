# Daten importieren
*Datenquelle: [https://www.kaggle.com/datasets/harlfoxem/housesalesprediction/data](https://www.kaggle.com/datasets/harlfoxem/housesalesprediction/data)*

```python
import pandas as pd

# Schritt 1: Lese die entsprechende Datei ein, speichere die Daten in der Variable 'df' ab und gib sie anschließend aus.
df = pd.read_csv("kc_house_data.csv")
df
```

```python
# Schritt 2: Analysiere die Daten mithilfe von '.describe()'

df.describe()
```

# Daten aufteilen

```python
# Schritt 3: Weise die entsprechenden Features der Eingabevariable X und das zugehörige Ziel der Zielvariable y zu.
# Tipp: Verwende '.copy()' und '.pop()'.

X = df.copy()
y = X.pop("price")
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

#Schritt 4: Unterteile die Daten in Trainings- und Testdaten, wobei die Testdaten 20% der Gesamtdaten ausmachen sollten.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
y_train
```

# Kreuzvalidierung

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

# Schritt 5: Erstelle eine Instanz eines Entscheidungsbaums mit der maximalen Tiefe von 3 und führe eine Kreuzvalidierung mit 4 Folds durch. Verwende hierfür die Metrik "neg_mean_squared_error".
cv_model = DecisionTreeRegressor(max_depth=3, random_state=0)
cross_val_scores = cross_val_score(cv_model, X_train, y_train, cv=4, scoring="neg_mean_squared_error")

print("Cross-Validation Scores: ", cross_val_scores)
```

```python
-1 * cross_val_scores
```

```python
# Schritt 6: Multipliziere die Werte in 'cross_val_scores' jeweils mit -1 und gib anschließend den Durchschnittsfehler aus.

positive_scores = -1 * cross_val_scores

print("Mean MSE: ", positive_scores.mean())
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

# Schritt7: Erstelle eine Instanz eines Entscheidungsbaums und definiere eine Liste mit verschiedenen Werten für 'max_depth'.
grid_cv_model = DecisionTreeRegressor(random_state=0)
param_grid = {
    'max_depth': [5, 6, 7, 8, 9, 10, 11]
}

# Schritt 8: Setze bei GridSearchCV die Anzahl an Folds auf 4 und scoring auf "neg_mean_squared_error".
grid_search = GridSearchCV(grid_cv_model, param_grid, cv=4, scoring="neg_mean_squared_error")
grid_search.fit(X_train, y_train)

print("Best Mean Squared Error: ", -grid_search.best_score_)
print("Best Parameters: ", grid_search.best_params_)
```

# Finales Modell

```python
from sklearn.metrics import mean_squared_error

# Schritt 9: Speichere das beste Modell in 'final_model' ab mithilfe von '.best_estimator_' und berechne den finalen MSE auf den Testdaten.
final_model = grid_search.best_estimator_
y_predict_final = final_model.predict(X_test)
mse = mean_squared_error(y_predict_final, y_test)

print("Final MSE:", mse)
```
