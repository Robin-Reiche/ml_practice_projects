# Daten importieren


```python
import pandas as pd

# Schritt 1: Lese die entsprechende Datei ein, speichere die Daten in der Variable 'df' ab und gib sie anschlie√üend aus.
df = pd.read_csv("purchasing_behavior_data.csv")
df
```

```python
# Schritt 2: Zeige die H√§ufigkeit jeder Klasse an, indem du '.value_counts()' verwendest.

df["Purchase_Made"].value_counts()
```

# Daten visualisieren

```python
import seaborn as sns

# Schritt 3: Visualisiere die Daten mithilfe eines Scatterplots.
sns.scatterplot(x=df["Minutes_Spent"], y=df["Purchase_Made"])
```

# Daten aufteilen

```python
# Schritt 4: Weise das entsprechende Feature der Eingabevariable X und das zugeh√∂rige Ziel der Zielvariable y zu.

X = df[["Minutes_Spent"]]
y = df["Purchase_Made"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

#Schritt 5: Unterteile die Daten in Trainings- und Testdaten, wobei die Testdaten 20% der Gesamtdaten ausmachen sollten.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
y_train
```

# Modell trainieren

```python
from sklearn.tree import DecisionTreeClassifier

# Schritt 6: Erstelle eine Instanz eines Entscheidungsbaums und trainiere dieses Modell auf den Trainingsdaten.
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

# Modell visualisieren

```python
from sklearn.tree import plot_tree

# Schritt 7: Visualisiere das Modell mithilfe von 'plot_tree()'.
plot_tree(model, feature_names=["Minutes_Spent"])
```

# Vorhersage

```python
# Schritt 8: Nutze das trainierte Modell, um Vorhersagen f√ºr die Testdaten zu generieren.

y_predict = model.predict(X_test)
```

# Modellbewertung (Accuracy)

```python
from sklearn.metrics import accuracy_score

# Schritt 9: Berechne die Accuracy basierend auf den Vorhersagen und den Testdaten und vervollst√§ndige die 'print()'-Funktion.
accuracy = accuracy_score(y_predict, y_test)
print("Accuracy:", accuracy)
```
